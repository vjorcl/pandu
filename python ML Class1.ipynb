```python
import pandas as pd        #to read and manuculte data set
import numpy as np        # used for numerical datsets
import matplotlib.pyplot as plt  #used for visualize data

```

```python

from sklearn.model_selection import train_test_split    #splits the datset into training and testing sets
from sklearn.linear_model import LinearRegression   #impements the linear regression algorithm
from sklearn.metrics import mean_squared_error,r2_score    #used to evalate the mode
```

```python
train=pd.read_csv("/content/Housing.csv")
print(train.head())
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    x=train['area'],
    y=train['price']
)
plt.xlabel=("Area")
plt.ylabel("Price")
plt.title("AREA VS Price")
plt.show()
```

```python
X=train[['area']]
Y=train['price']

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)

#train model
model=LinearRegression()
model.fit(x_train,y_train)
print("Slope:",model.coef_[0])
print("Intersept",model.intercept_)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    x_train,y_train,alpha=0.5,label="Actual Data"
)

#regression line
plt.plot(x_train,
         model.predict(x_train),
         color='red',
         label="Regression Line"

         )
plt.xlabel=("Area")
plt.ylabel("Price")
plt.title("Area Vs Price")
plt.legend()
plt.show()
```
